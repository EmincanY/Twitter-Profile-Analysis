```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re

# Start the Selenium driver
driver = webdriver.Chrome()

# Go to the user profile
driver.get("https://x.com/tarihiiimerak")

# Scrolling parameters
scroll_pause_time = 2  # Waiting time after each scroll (in seconds)
scroll_count = 15       # Number of times to scroll
tweet_limit = 50       # Number of tweets to fetch

# Variables for storing tweets
tweet_texts = []
tweet_ids = set()  # To prevent duplicate tweets
collect_tweet = False

# Perform the scrolling process
for _ in range(scroll_count):
    # Retrieve the page source after each scroll
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    tweets = soup.find_all('span', {'class': 'css-1jxf684'})
    
    current_tweet = []

    for tweet in tweets:
        text = tweet.get_text(strip=True)

        # Start collecting after detecting two consecutive dots and user handle
        if text == "·" and not collect_tweet:
            collect_tweet = True
            continue

        # Finish the current tweet and save it when a new @tarihiiimerak handle is detected
        if text == "@tarihiiimerak" and collect_tweet:
            if current_tweet:
                full_tweet = " ".join(current_tweet).strip()

                # Use regex to clean up repeated two-digit sequences and everything after them
                cleaned_tweet = re.sub(r'\b(\d+)\b\s+\b\1\b.*', '', full_tweet)
                
                # Add new tweet if it's unique
                tweet_id = hash(cleaned_tweet)  # Create a unique ID for the tweet
                if tweet_id not in tweet_ids:
                    tweet_texts.append(cleaned_tweet)
                    tweet_ids.add(tweet_id)  # Save the tweet ID

                current_tweet = []

                # Stop if the tweet limit is reached
                if len(tweet_texts) >= tweet_limit:
                    break
            continue

        # Collect tweet content
        if collect_tweet and text != "@tarihiiimerak":
            current_tweet.append(text)

    # Stop scrolling if the tweet limit is reached
    if len(tweet_texts) >= tweet_limit:
        break

    # Scroll a bit further down the page
    driver.execute_script("window.scrollBy(0, 3000);")  # Scrolls down 3000 pixels
    time.sleep(scroll_pause_time)

# Print the collected tweets
for idx, tweet_text in enumerate(tweet_texts, 1):
    print(f"Tweet {idx}:\n{tweet_text}\n")

# Close the driver
driver.quit()
```
